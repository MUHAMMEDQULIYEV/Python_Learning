```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_theme()
iris=sns.load_dataset("iris")
iris
```

```python
iris['species'].unique()
```

```python
iris['species'].nunique()
```

```python
sns.pairplot(data=iris)
plt.savefig("iris_data.png")
```

```python
g=sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g.map_diag(sns.displot)
```

```python
g.map_lower(sns.kdeplot)
```
