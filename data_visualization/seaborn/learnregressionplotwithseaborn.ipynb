```python
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset("tips")
%matplotlib inline
```

```python
tips
```

```python
sns.lmplot(data=tips,x='total_bill',y="tip")
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue="sex")
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex',markers=['o','v'])
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex',markers=['o','v'],scatter_kws={"s":100})
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',col='day',row='time',hue='sex',markers=['o','v'],aspect=0.6)
```
