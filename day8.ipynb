```python
# Index Levels
outside = ['G1','G1','G1','G2','G2','G2']
inside = [1,2,3,1,2,3]
hier_index = list(zip(outside,inside))
hier_index=pd.MultiIndex.from_tuples(hier_index)
```

```python
fd=pd.DataFrame(np.random.randn(6,2),index=hier_index,columns=['A','B'])
fd.index = fd.index.set_levels(fd.index.levels[1].astype(int), level=1)
```
